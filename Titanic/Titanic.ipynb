## Introduction

We'll be working with a data set of the passengers of the Titanic. The Titanic shipwreck is the most famous shipwreck in history and led to the creation of better safety regulations for ships. One substantial safety issue was that there were not enough lifeboats for every passenger on board, which meant that some passengers were prioritized over others to use the lifeboats.

The data set was compiled by Kaggle for their introductory data science competition, called Titanic: Machine Learning from Disaster. The goal of the competition is to build machine learning models that can predict if a passenger survives from their attributes.

Here is a preview of the dataset:

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

titanic = pd.read_csv('train.csv')
titanic.head(4)
```

 - PassengerId -- A numerical id assigned to each passenger.
 - Survived -- Whether the passenger survived (1), or didn't (0).
 - Pclass -- The class the passenger was in.
 - Name -- the name of the passenger.
 - Sex -- The gender of the passenger -- male or female.
 - Age -- The age of the passenger. Fractional.
 - SibSp -- The number of siblings and spouses the passenger had on board.
 - Parch -- The number of parents and children the passenger had on board.
 - Ticket -- The ticket number of the passenger.
 - Fare -- How much the passenger paid for the ticker.
 - Cabin -- Which cabin the passenger was in.
 - Embarked -- Where the passenger boarded the Titanic.
 
Let's remove columns like Name and Ticket that we don't have a way to visualize. In addition, we need to remove any rows containing missing values to avoid any errors while plotting.

```python
cols = ["Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]
titanic = titanic[cols].dropna(axis=0)
titanic.head(4)
```

## Distribution of Passengers' Age

```python
sns.distplot(titanic["Age"])
plt.show()
```

## Kernal Density Plot of Passengers' Age

```python
sns.set_style("white")
sns.kdeplot(titanic["Age"], shade=True)
plt.xlabel("Age")
sns.despine(left=True, bottom=True)
plt.show()
```

## Kernal Density Plots of Passengers' Age by Ticket Class

```python
g = sns.FacetGrid(titanic, col="Pclass", size=6)
g.map(sns.kdeplot, "Age", shade=True)
sns.despine(left=True, bottom=True)
plt.show()
```

## Kernal Density Plots of Passengers' Age by Ticket Class and Survival

```python
h = sns.FacetGrid(titanic, col="Pclass", row="Survived")
h.map(sns.kdeplot, "Age", shade=True)
sns.despine(left=True, bottom=True)
plt.show()
```

## Kernal Density Plots of Passengers' Age by Ticket Class, Survival and Sex

```python
i = sns.FacetGrid(titanic, col="Pclass", row="Survived", hue="Sex", size=4)
i.map(sns.kdeplot, "Age", shade=True).add_legend()
sns.despine(left=True, bottom=True)
plt.show()
```
