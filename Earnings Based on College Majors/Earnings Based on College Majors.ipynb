# Introduction

We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more.

Using visualizations, we can start to explore questions from the dataset like:
 - Do students in more popular majors make more money?
    -  Using scatter plots
 - How many majors are predominantly male? Predominantly female?
    -  Using histograms
 - Which category of majors have the most students?
    -  Using bar plots

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
raw_recent_grads = pd.read_csv("recent-grads.csv")

raw_recent_grads.iloc[0]
```

```python
raw_recent_grads.head()
```

```python
raw_recent_grads.tail()
```

```python
raw_recent_grads.describe()
```

```python
raw_dimensions = raw_recent_grads.shape
raw_data_count = raw_dimensions[0]

recent_grads = raw_recent_grads.dropna()
dimensions = recent_grads.shape
cleaned_data_count = dimensions[0]

print(raw_data_count)
print(cleaned_data_count)
```

## Scatter Plots
### Sample Size vs. Median Salary of Full-Time Employed Graduates

```python
ax1 = recent_grads.plot(x = 'Sample_size', y = 'Median', kind = 'scatter')
ax1.set_title('Sample Size vs. Median Salary')
```

The graph shows a decreasing trend indicating that students studying less popular majors make more money.

### Sample Size of Full-Time Employed Graduates vs. Unemployment Rate

```python
ax2 = recent_grads.plot(x = 'Sample_size', y = 'Unemployment_rate', kind = 'scatter')
ax2.set_title('Sample Size vs. Unemployment Rate')
```

### Full Time Employed Graduates vs. Median Salary

```python
ax3 = recent_grads.plot(x = 'Full_time', y = 'Median', kind = 'scatter')
ax3.set_title('Full Time vs. Median Salary')
```

### Share of Female Graduates vs. Unemployment Rate

```python
ax4 = recent_grads.plot(x = 'ShareWomen', y = 'Unemployment_rate', kind = 'scatter')
ax4.set_title('Share of Female Graduates vs. Unemployment Rate')
```

### Male Graduates vs. Median Salary of Full Time Employed Graduates

```python
ax5 = recent_grads.plot(x = 'Men', y = 'Median', kind = 'scatter')
ax5.set_title('Male Graduates vs. Median Salary')
```

### Female Graduates vs. Median Salary of Full Time Employed Graduates

```python
ax5 = recent_grads.plot(x = 'Women', y = 'Median', kind = 'scatter')
ax5.set_title('Female Graduates vs. Median Salary')
```
